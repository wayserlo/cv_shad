```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
from skimage.color import gray2rgb
```

```python
a = np.load("original_img.npy")
b = np.load("blurred_img.npy")
k = np.load("kernel.npy")
```

```python
c = scipy.signal.convolve2d(a, k, mode="same", boundary="wrap")
delta = 1
noise = np.random.randint(-delta, 1+delta, size=c.shape)/255
noise -= noise.mean()
c += noise
print(c.min(), c.max())
c = c.clip(0, 1)
np.save("blurred_img.npy", c)
```
